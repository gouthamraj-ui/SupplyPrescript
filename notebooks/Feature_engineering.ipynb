```python
#Lead Time categroy
def lead_category(days):

    if days <= 5:
        return "Fast"

    elif days <= 10:
        return "Average"

    else:
        return "Slow"

df["lead_time_category"] = (
    df["lead_time"]
      .apply(lead_category)
)
```

```python
#Cost Impact category
def cost_category(cost):

    if pd.isna(cost):
        return "No Cost"

    elif cost < 2000:
        return "Low"

    elif cost < 5000:
        return "Medium"

    else:
        return "High"

df["cost_impact"] = (
    df["cost_change"]
      .apply(cost_category)
)
```

```python
#Supplier performance
supplier_delay = (
    df.groupby("supplier_name")["delivery_delay"]
      .mean()
)

df["supplier_avg_delay"] = (
    df["supplier_name"]
      .map(supplier_delay)
)
```

```python
#Inventory status
def inventory_status(stock):

    if stock < 100:
        return "Low"

    elif stock < 300:
        return "Medium"

    else:
        return "High"

df["inventory_status"] = (
    df["stock_quantity"]
    .apply(inventory_status)
)
```

```python
#Large shipment
average_qty = df["shipment_quantity"].mean()

df["large_shipment"] = (
    df["shipment_quantity"] > average_qty
).astype(int)
```

```python
#High severity Disruption
df["high_severity"] = (
    df["severity"] == "High"
).astype(int)
```

```python
#Delay Category
def delay_category(days):

    if days <= 0:
        return "On Time"

    elif days <= 3:
        return "Minor Delay"

    else:
        return "Major Delay"

df["delay_category"] = df["delivery_delay"].apply(delay_category)
```

```python
#Is Delayed
df["is_delayed"] = (
    df["delivery_delay"] > 0
).astype(int)
```

```python
#shipment Weekday
df["shipment_weekday"] = df["shipment_date"].dt.day_name()
```

```python
#shipment Quarter
df["shipment_quarter"] = df["shipment_date"].dt.quarter
```

```python
#shipment year
df["shipment_year"] = df["shipment_date"].dt.year
```

```python
# Shipment month
df["shipment_month"] = df["shipment_date"].dt.month_name()
```

```python
#Delivery Delay
df["delivery_delay"] = (
    df["actual_arrival"] -
    df["expected_arrival"]
).dt.days
```
