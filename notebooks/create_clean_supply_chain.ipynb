```python
import pandas as pd

# Read datasets
suppliers = pd.read_csv("../data/suppliers.csv")
products = pd.read_csv("../data/products.csv")
inventory = pd.read_csv("../data/inventory.csv")
shipments = pd.read_csv("../data/shipments_cleaned.csv")
disruptions = pd.read_csv("../data/disruptions.csv")
decisions = pd.read_csv("../data/decisions.csv")
outcomes = pd.read_csv("../data/outcomes.csv")

# Rename columns
shipments = shipments.rename(columns={"quantity": "shipment_quantity"})
products = products.rename(columns={"supplier_id": "product_supplier_id"})

# Start with shipments
df = shipments.copy()

# Merge products
df = df.merge(products, on="product_id", how="left")

# Merge suppliers using shipment supplier_id
df = df.merge(suppliers, on="supplier_id", how="left")

# Merge inventory
df = df.merge(inventory, on="product_id", how="left")

# Merge disruptions
df = df.merge(disruptions, on="shipment_id", how="left")

# Merge decisions
df = df.merge(decisions, on="disruption_id", how="left")

# Merge outcomes
df = df.merge(outcomes, on="decision_id", how="left")

print(df.head())
```
